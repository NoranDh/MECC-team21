# Process

```python
# Define a TypedDict to store the agent's state
class AgentState(TypedDict):
    start: bool             # Indicates if the conversation has started
    conversation: int       # Keeps track of conversation turns
    question: str           # Customer's question
    answer: str             # Agent's answer
    topic: bool             # Decision on whether the question is appropriate
    documents: list         # Relevant documents retrieved
    recursion_limit: int    # To avoid endless loops
    memory: list            # Conversation history
```

```python
def greetings(state):
    # Greet the student
    print("Hello! Welcome. I am Failure analysis. How can I help you?")

    # Capture user input
    user_input = input()

    # Update the state with the student's question and initialize conversation variables
    state['question'] = user_input
    state['conversation'] = 1
    state['memory'] = [user_input]

    return state
```

```python
def check_question(state):
    # Get the student's question from the state
    question = state['question']

    # Define the system prompt to evaluate the question's appropriateness
    system_prompt = """
You are a helpful agent that extracts matertials failure analyzer information from the file that you are given.
    Read the information accuratly and extract each problem with its relevant information.
    Extract a realistic material filure analysis data that when you receive it you will understand the problem, sequence of events, boundary conditions (add to it additional note as the one in the example below).

Return only the relevant information in JSON format.
If any information is missing, return its value as "unknown".

Your answer should be like the following example:
{
  "failure_cause": "The fatal explosion and fire were caused by the release of flammable solvent vapor through the seal of a closed reactor manway. The manway was not designed, constructed, or pressure tested to a proper design pressure for the process.",
  "sequence_of_events": "The reactor became pressurized by rapid solvent vaporization when the agitator was turned on, which caused the release through the closed manway. The absence of engineering controls and reliance on operator actions contributed to the failure.",
  "boundary_conditions": {
    "pressure_handling_capabilities": "The manway's sealing surface began to release flammable material when pressure exceeded 9 psig, though the rupture disk's documented burst pressure was 15 psig.",
    "pressure_testing": "The manway was only leak-checked up to 4 psig during installation and was not adequately pressure tested to withstand expected conditions.",
    "Additional note": "It should be noted that the manway failure occurred at a pressure lower than the rupture disk set point (9 psig vs. 15 psig), indicating that the manway was the weakest point in the system and failed before the overpressure protection could function as intended."
  }
}
Explanation:

The information about the failure's cause and sequence of events primarily involves the improper design and testing of the reactor manway and how it failed to handle the pressure caused by the rapid vaporization when the agitator was turned on.
Boundary conditions include the failure to adequately test the manway and prepare for expected pressure conditions.

    """

    # Create a prompt template for formatting
    TEMPLATE = ChatPromptTemplate.from_messages([
        ("system", system_prompt),
        ("human", "User question: {question}"),
    ])

    # Format the prompt with the student's question
    prompt = TEMPLATE.format(question=question)

    # Initialize the language model
    model = ChatOpenAI(model="gpt-4o-mini", api_key=api_key)

    # Invoke the model with the prompt
    response_text = model.invoke(prompt)

    # Update the state with the model's decision
    state['topic'] = response_text.content.strip()

    return state
```

```python
# Function to check if the LLM model decides if the question is on topic on top
def topic_router(state):
  topic = state['topic']
  if topic == "True":
    return "on_topic"
  else:
    return "off_topic"
```

```python
def off_topic_response(state):
    # Check if it's the first conversation turn
    if state['conversation'] <= 1:
        state['answer'] = "\nI apologize, I can't answer that question. I can only answer questions about the AOU academic related."
        print(state['answer'])
    else:
        state['answer'] = "\nHappy to help."
        print(state['answer'])
```

```python
def retrieve_docs(state):
    # Combine the conversation history into a single string
    memory = " ".join(state['memory'])

    # Retrieve the top 5 relevant documents based on similarity to the conversation history
    docs_faiss = db.similarity_search(memory, k=5)

    # Store the retrieved documents' content in the state
    state['documents'] = [doc.page_content for doc in docs_faiss]

    return state
```

```python
def generate(state):
    # Initialize the language model
    model = ChatOpenAI(model="gpt-4o-mini", api_key=api_key)

    # Extract necessary information from the state
    question = state['question']
    documents = state['documents']
    memory = state['memory']

    # Define the system prompt
    system_prompt = """
    You are a matirial failure analyzer tasked with extracting the matirial failure analysis information.
    Extract a realistic filure analysis data that when you receive it you will understand the problem, sequence of events, boundary conditions (add to it additional note as the one in the example below).
    Return only the relevant information in JSON format.
"""
    # Create a prompt template for formatting
    TEMPLATE = ChatPromptTemplate.from_messages([
        ("system", system_prompt),
        ("human", "Context: {documents}\nConversation history so far: {memory}\nanalysis question: {question}"),
    ])

    # Format the prompt with the retrieved documents, conversation history, and student's question
    prompt = TEMPLATE.format(documents=documents, memory=memory, question=question)

    # Invoke the model to generate an answer
    response_text = model.invoke(prompt)

    # Store the generated answer in the state
    state["answer"] = response_text.content.strip()

    return state
```

```python
# Define the system prompt for refining the answer
system_prompt = """
You are a helpful agent that extracts matertials failure analyzer information from the file that you are given.
    Read the information accuratly and extract each problem with its relevant information.
    Extract a realistic material filure analysis data that when you receive it you will understand the problem, sequence of events, boundary conditions (add to it additional note as the one in the example below).

Return only the relevant information in JSON format.
If any information is missing, return its value as "unknown".

Your answer should be like the following example:
{
  "failure_cause": "The fatal explosion and fire were caused by the release of flammable solvent vapor through the seal of a closed reactor manway. The manway was not designed, constructed, or pressure tested to a proper design pressure for the process.",
  "sequence_of_events": "The reactor became pressurized by rapid solvent vaporization when the agitator was turned on, which caused the release through the closed manway. The absence of engineering controls and reliance on operator actions contributed to the failure.",
  "boundary_conditions": {
    "pressure_handling_capabilities": "The manway's sealing surface began to release flammable material when pressure exceeded 9 psig, though the rupture disk's documented burst pressure was 15 psig.",
    "pressure_testing": "The manway was only leak-checked up to 4 psig during installation and was not adequately pressure tested to withstand expected conditions.",
    "Additional note": "It should be noted that the manway failure occurred at a pressure lower than the rupture disk set point (9 psig vs. 15 psig), indicating that the manway was the weakest point in the system and failed before the overpressure protection could function as intended."
  }
}
Explanation:

The information about the failure's cause and sequence of events primarily involves the improper design and testing of the reactor manway and how it failed to handle the pressure caused by the rapid vaporization when the agitator was turned on.
Boundary conditions include the failure to adequately test the manway and prepare for expected pressure conditions.
    """
```

```python
def improve_answer(state):
    # Extract necessary information from the state
    question = state['question']
    answer = state['answer']
    memory = state['memory']

    # Define the system prompt for refining the answer
    system = system_prompt

    # Create a prompt template for formatting
    TEMPLATE = ChatPromptTemplate.from_messages([
        ("system", system),
        ("human", "analysis question: {question}\nConversation history: {memory}\nmatirial failure analyser initial answer: {answer}"),
    ])

    # Initialize the language model
    model = ChatOpenAI(model="gpt-4o-mini", api_key=api_key)

    # Format the prompt with the necessary information
    prompt = TEMPLATE.format(question=question, memory=memory, answer=answer)

    # Invoke the model to improve the answer
    response_text = model.invoke(prompt)

    # Update the state with the improved answer
    state['answer'] = response_text.content.strip()

    # Display the improved answer
    print('\n')
    print(state['answer'])

    # Append the improved answer to the conversation history
    state['memory'].append(state['answer'])

    return state
```

```python
def further_question(state):
    # Prompt the student for further input
    print('\n')
    user_input = input()

    # Update the state with the new question and increment the conversation turn
    state['question'] = user_input
    state['conversation'] += 1
    state['memory'].append(user_input)

    return state
```

# Workflow

```python
# Initialize a StateGraph with the AgentState type
workflow = StateGraph(AgentState)


# Add the functions as nodes in the workflow
workflow.add_node("greetings", greetings)
workflow.add_node("check_question", check_question)
workflow.add_node("off_topic_response", off_topic_response)
workflow.add_node("retrieve_docs", retrieve_docs)
workflow.add_node("generate", generate)
workflow.add_node("improve_answer", improve_answer)
workflow.add_node("further_question", further_question)

# Set the entry point of the workflow to the greetings function
workflow.set_entry_point("greetings")

# Add conditional edges based on the topic_router function's output
workflow.add_conditional_edges(
    "check_question",
    topic_router,
    {
        "on_topic": "retrieve_docs",
        "off_topic": "off_topic_response"
    }
)

# Define the sequence of steps in the workflow
workflow.add_edge("greetings", "check_question")
workflow.add_edge("retrieve_docs", "generate")
workflow.add_edge("generate", "improve_answer")
workflow.add_edge("improve_answer", "further_question")
workflow.add_edge("further_question", "check_question")

# Connect the off_topic_response node to the end of the workflow
workflow.add_edge("off_topic_response", END)

# Compile the workflow into an application
app = workflow.compile()
```

```python
# Display the workflow graph for visualization
display(Image(app.get_graph(xray=True).draw_mermaid_png()))
```

```python
# Invoke the application with the initial state
result = app.invoke({"start": True}, {"recursion_limit": 50})
```

```python
from datasets import Dataset
from ragas.testset import TestsetGenerator
from ragas import evaluate
import json
import pandas as pd
import numpy as np
import asyncio
from tqdm.auto import tqdm

# RAGAS imports (mirror your RAGAS notebook)
from ragas.dataset_schema import SingleTurnSample
from ragas.llms import LangchainLLMWrapper
from ragas.embeddings import LangchainEmbeddingsWrapper

from ragas.metrics import RougeScore, SimpleCriteriaScore, RubricsScore, SemanticSimilarity
from ragas.metrics import LLMContextPrecisionWithReference, LLMContextRecall, ResponseRelevancy
from ragas.metrics._factual_correctness import FactualCorrectness

# LangChain/OpenAI wrappers used in ragas notebook
from langchain_openai import ChatOpenAI
from langchain.embeddings import OpenAIEmbeddings
```

```python
test_file = "fine-tuning_eval_samples.jsonl"
use_improve_step = True
max_examples = None

# Load JSONL test file
test_records = []
with open(test_file, "r", encoding="utf-8") as f:
    for line in f:
        line = line.strip()
        if not line:
            continue
        test_records.append(json.loads(line))


df = pd.DataFrame(test_records)
if "input" in df.columns:
    df["user_input"] = df["input"]
else:
    df["user_input"] = df.iloc[:, 0].astype(str)

if "ideal" in df.columns:
    df["reference"] = df["ideal"]
else:
    # fallback: try second column
    df["reference"] = df.iloc[:, 1].astype(str)

if max_examples:
    df = df.iloc[:max_examples].reset_index(drop=True)
```

```python
# Define a non-interactive wrapper that calls your pipeline functions directly

def chatbot_call(question, use_improve=use_improve_step):
    """
    Build state, call retrieve_docs -> generate -> (optionally) improve_answer,
    and return (answer_text, retrieved_documents_list).
    """
    # Build initial state to match AgentState TypedDict
    state = {
        "start": False,
        "conversation": 1,
        "question": str(question),
        "answer": "",
        "topic": True,
        "documents": [],
        "recursion_limit": 50,
        "memory": [str(question)]
    }

    state = retrieve_docs(state)
    state = generate(state)

    # call improve step
    if use_improve:
        try:
            state = improve_answer(state)
        except Exception as e:
            # if improve fails, keep the original generated answer
            print("improve_answer failed for a sample (continuing):", str(e))

    answer = state.get("answer", "")
    documents = state.get("documents", [])
    # if they are doc objects with 'page_content', normalize:
    normalized_docs = []
    for d in documents:
        if isinstance(d, dict):
            content = d.get("page_content") or d.get("content") or str(d)
            normalized_docs.append(content)
        else:
            normalized_docs.append(str(d))
    return answer, normalized_docs

```

```python
# Run chatbot on each test input and collect outputs
predictions = []
contexts_list = []
for q in tqdm(df["user_input"].tolist(), desc="Generating answers"):
    ans, docs = chatbot_call(q, use_improve=use_improve_step)
    predictions.append(ans)
    contexts_list.append(docs)

# Build the RAGAS-style DataFrame expected by the RAGAS notebook
test_df = pd.DataFrame({
    "user_input": df["user_input"],
    "reference": df["reference"],
    "answer": predictions,
    "context": contexts_list
})
```

```python
# Prepare RAGAS scorers (LLM + Embeddings wrappers)
api_kw = {}
if "api_key" in globals():
    api_kw['api_key'] = globals().get("api_key")

# Evaluation LLM + embeddings
eval_llm = LangchainLLMWrapper(ChatOpenAI(model="gpt-4.1", temperature=0, **api_kw))
# For embeddings, pass same API key if available
if "api_key" in globals():
    eval_embeddings = LangchainEmbeddingsWrapper(OpenAIEmbeddings(openai_api_key=globals().get("api_key")))
else:
    eval_embeddings = LangchainEmbeddingsWrapper(OpenAIEmbeddings())

# Instantiate scorers
rouge_scorer = RougeScore()
simple_scorer = SimpleCriteriaScore(name="simple scorer", definition="Score 0 to 5 by similarity", llm=eval_llm)
# Rubrics require a rubrics dict.
rubrics = globals().get("rubrics", {
    "score1_description": "The response is entirely incorrect and fails to address any aspect of the reference.",
    "score2_description": "The response contains partial accuracy but includes major errors or significant omissions that affect its relevance to the reference.",
    "score3_description": "The response is mostly accurate but lacks clarity, thoroughness, or minor details needed to fully address the reference.",
    "score4_description": "The response is accurate and clear, with only minor omissions or slight inaccuracies in addressing the reference.",
    "score5_description": "The response is completely accurate, clear, and thoroughly addresses the reference without any errors or omissions."
})
rubrics_scorer = RubricsScore(rubrics=rubrics, llm=eval_llm)

semantic_scorer = SemanticSimilarity(embeddings=eval_embeddings)
context_precision_scorer = LLMContextPrecisionWithReference(llm=eval_llm)
context_recall_scorer = LLMContextRecall(llm=eval_llm)
response_relevancy_scorer = ResponseRelevancy(llm=eval_llm, embeddings=eval_embeddings)
factual_scorer = FactualCorrectness(llm=eval_llm)

# Helper to normalize returned scores (some scorers return dicts)
def _num_from_score(s):
    if s is None:
        return None
    if isinstance(s, (int, float, np.number)):
        return float(s)
    if isinstance(s, dict):
        # common keys
        for k in ("score", "value", "rating"):
            if k in s:
                try:
                    return float(s[k])
                except:
                    pass
        # fallback: try to find a numeric entry
        for v in s.values():
            if isinstance(v, (int, float, np.number)):
                return float(v)
        return None
    # else try convert
    try:
        return float(s)
    except:
        return None
```

```python
# Async evaluation loop that calls
async def _evaluate_async(df_in):
    # result lists
    results = {
        "rouge": [],
        "simple": [],
        "rubrics": [],
        "semantic": [],
        "context_precision": [],
        "context_recall": [],
        "response_relevancy": [],
        "factual": []
    }

    records = df_in.to_dict(orient="records")
    for r in tqdm(records, desc="Scoring (RAGAS metrics)"):
        sample = SingleTurnSample(
            user_input = r["user_input"],
            reference = r["reference"],
            response = r["answer"],
            retrieved_contexts = r.get("context", [])
        )

        # For each scorer, await its single_turn_ascore
        try:
            val = await rouge_scorer.single_turn_ascore(sample)
            results["rouge"].append(_num_from_score(val))
        except Exception as e:
            print("Rouge scoring error:", e); results["rouge"].append(None)

        try:
            val = await simple_scorer.single_turn_ascore(sample)
            results["simple"].append(_num_from_score(val))
        except Exception as e:
            print("Simple scoring error:", e); results["simple"].append(None)

        try:
            val = await rubrics_scorer.single_turn_ascore(sample)
            results["rubrics"].append(_num_from_score(val))
        except Exception as e:
            print("Rubrics scoring error:", e); results["rubrics"].append(None)

        try:
            val = await semantic_scorer.single_turn_ascore(sample)
            results["semantic"].append(_num_from_score(val))
        except Exception as e:
            print("Semantic scoring error:", e); results["semantic"].append(None)

        try:
            val = await context_precision_scorer.single_turn_ascore(sample)
            results["context_precision"].append(_num_from_score(val))
        except Exception as e:
            print("Context precision error:", e); results["context_precision"].append(None)

        try:
            val = await context_recall_scorer.single_turn_ascore(sample)
            results["context_recall"].append(_num_from_score(val))
        except Exception as e:
            print("Context recall error:", e); results["context_recall"].append(None)

        try:
            val = await response_relevancy_scorer.single_turn_ascore(sample)
            results["response_relevancy"].append(_num_from_score(val))
        except Exception as e:
            print("Response relevancy error:", e); results["response_relevancy"].append(None)

        try:
            val = await factual_scorer.single_turn_ascore(sample)
            results["factual"].append(_num_from_score(val))
        except Exception as e:
            print("Factual scoring error:", e); results["factual"].append(None)

    return results

# Run async scorer
scorer_results = asyncio.run(_evaluate_async(test_df))
```

```python
# Aggregate and print results
def mean_ignore_none(lst):
    vals = [v for v in lst if v is not None]
    return float(np.mean(vals)) if len(vals)>0 else None

agg = {k: mean_ignore_none(v) for k,v in scorer_results.items()}

print("\nRAGAS evaluation results (means across dataset):")
for k,v in agg.items():
    print(f"{k:20s} : {v}")

# Also attach per-example scores into test_df (optinal)
for k,v in scorer_results.items():
    test_df[f"score_{k}"] = v

# Save results
test_df.to_csv("ragas_evaluation_results.csv", index=False)
print("\nPer-example scores saved to ragas_evaluation_results.csv")
```
